```python
import pandas as pd
```

```python
df=pd.read_csv("./Resources/cities_data.csv")
df.head()
```

```python
df=df.dropna().reset_index()
```

```python
df.head(35)
```

```python
df=df.drop(['index'], axis=1)
```

```python
df = df.rename_axis("City_ID")
```

```python
df.head()
```

```python
df=df.rename(columns={'city': 'City', 'lat': 'Lat', 'long': 'Long', 'temp': 'Max Temp', 'humidity': 'Humidity', 
                      'cloudiness': 'Cloudiness', 'windspeed': 'Wind Speed', 'country': 'Country', 'date': 'Date'})
df.head()
df.tail()
```

```python
column_names=['City', 'Cloudiness', 'Country', 'Date', 'Humidity', 'Lat','Long', 'Max Temp', 'Wind Speed']
df.head(35)
```

```python
df = df.reindex(columns=column_names)
```
